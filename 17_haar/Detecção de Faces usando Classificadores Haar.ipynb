## Aula 17 - Detecção de Faces usando Classificadores Haar **(parte 1)**

Começamos importando as bibliotecas necessárias:

```python
%matplotlib inline
import cv2 # OpenCV
import matplotlib.pyplot as plt # Matplotlib
import numpy as np # Numpy
```

Vamos estudar os básicos da detecção de faces usando os classificadores Haar. Usaremos o mesmo conhecimento para detectar os olhos das pessoas, mas o pipeline pode ser seguido para qualquer base do Haar!

```python
img = plt.imread('mila.jpg')
plt.imshow(img)
```

Os classificadores são salvos em um formato XML. Vamos carregá-los:

```python
face_cascade = cv2.CascadeClassifier('class/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('class/haarcascade_eye.xml')
```

Precisamos converter a imagem para tons de cinza:

```python
cinza = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(cinza, cmap="gray")
```

Com o classificador carregado, encontrar os rostos nas imagens é fácil:

```python
faces = face_cascade.detectMultiScale(cinza, scaleFactor=1.3, minNeighbors=5)
```

Quantas faces temos na imagem?

```python
len(faces)
```

*Faces* é um vetor **(n, 4)**, onde **n** é o número de faces com 4 informações necessárias para desenhar o *bounding-box*:
1. x;
2. y;
3. w;
4. h;

Podemos usar isso para desenhar um quadrado ao redor do rosto na imagem:

```python
img_c = img.copy()
for (x,y,w,h) in faces:
    cv2.rectangle(img_c,(x,y),(x+w,y+h),(255,0,0),8)
    plt.imshow(img_c)
```

A estratégia para detectar os olhos é a mesma, porém usamos a informação do rosto para reduzir a quantidade de informações na imagem. Logo, iremos buscar os olhos apenas na região pré-detectada do rosto, que é onde os olhos (normalmente) estão.

**Observação**: por um bug no pyplot, rode a célula abaixo duas vezes. :)

```python
img_c = img.copy()
for (x,y,w,h) in faces:
    cv2.rectangle(img_c,(x,y),(x+w,y+h),(255,0,0),8)
    
    # rdi: região de interesse
    rdi = cinza[y:y+h, x:x+w]
    rdi_cor = img[y:y+h, x:x+w]
    olhos = eye_cascade.detectMultiScale(rdi)
    for (ox,oy,ow,oh) in olhos:
        cv2.rectangle(rdi_cor,(ox,oy),(ox+ow,oy+oh),(0,255,0),2)
plt.imshow(img_c)
```

Vamos tentar com uma outra imagem:

```python
img = plt.imread('stark.jpg')
cinza = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
faces = face_cascade.detectMultiScale(cinza, scaleFactor=1.3, minNeighbors=5)
```

```python
img_c = img.copy()
for (x,y,w,h) in faces:
    cv2.rectangle(img_c,(x,y),(x+w,y+h),(255,0,0),8)
    
    # rdi: região de interesse
    rdi = cinza[y:y+h, x:x+w]
    rdi_cor = img[y:y+h, x:x+w]
    olhos = eye_cascade.detectMultiScale(rdi, scaleFactor=1.3, minNeighbors=5)
    for (ox,oy,ow,oh) in olhos:
        cv2.rectangle(rdi_cor,(ox,oy),(ox+ow,oy+oh),(0,255,0),2)
plt.imshow(img_c)
print('Olhos detectados: ', len(olhos))
```

Mudem os parâmetros **scaleFactor** e **minNeighbors** e acompanhem a quantidade de olhos que é detectada na imagem.

Com o classificador treinado, podemos carregá-lo e testar na nossa imagem:
