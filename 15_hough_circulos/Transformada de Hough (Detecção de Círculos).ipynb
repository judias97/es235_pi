## Aula 15 - Transformada de Hough - Detecção de Círculos

Vamos aplicar a Transformada de Hough usando o OpenCV para **detecção de círculos**.

Começamos importando as bibliotecas necessárias:

```python
%matplotlib inline
import cv2 # OpenCV
import matplotlib.pyplot as plt # Matplotlib
import numpy as np # Numpy
```

Agora, carregaremos a imagem de testes. Usaremos uma imagem com um círculo bem visível.

```python
img = cv2.imread('240_pepsi_12oz.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_cinza = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img)
```

Algumas implementações desse processo iniciam suavizando a imagem com um filtro gaussiano ou de médias. Iremos seguir essa ideia, mas fica a sugestão: **testar novamente sem o filtro.**

```python
# Comente a linha abaixo para fazer o processo sem suavização
img = cv2.medianBlur(img,5)
```

Já existe uma implementação para a detecção de círculos. Usaremos o método [cv2.HoughCircles](https://docs.opencv.org/3.1.0/dd/d1a/group__imgproc__feature.html#ga47849c3be0d0406ad3ca45db65a25d2d) para isso, e é recomendável ler a documentação para entender alguns parâmetros.

Aqui iremos *tunar* dois parâmetros:

1. **dp** - a taxa inversa da resolução do acumulador em relação à resolução da imagem. Se dp=1, o acumulador tem a mesma resolução da imagem original, se dp=2, metade, e etc.

2. **minDist** - a distância mínima entre os centros dos círculos detectados. Se for muito grande, alguns círculos são ignorados.

```python
circulos = cv2.HoughCircles(img_cinza, cv2.HOUGH_GRADIENT, dp=1.2, minDist=100)
```

A variável **circulos** agora é uma lista de listas contendo a seguinte informação de cada círculo:
1. x
2. y
3. r (raio)

Vamos verificar:

```python
if circulos is not None:
    circulos = np.round(circulos[0, :]).astype("int")
    for circulo in circulos:
        print('(%d, %d) com raio %d' % (circulo[0], circulo[1], circulo[2]))
```

Vamos desenhar cada círculo que detectamos na imagem original:

```python
for (x, y, r) in circulos:
    cv2.circle(img, (x, y), r, (255, 0, 0), 4)
plt.imshow(img)
```

Vamos partir para um caso mais desafiador:

```python
img = cv2.imread('plot.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_cinza = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img)
```

Quando usamos a suavização na imagem da lata da Pepsi tivemos pouca mudança no resultado. Que tal testar aqui?

```python
# Comente a linha abaixo para fazer o processo sem suavização
img_blur = cv2.medianBlur(img,7)
img_cinza_blur = cv2.cvtColor(img_blur, cv2.COLOR_RGB2GRAY)

fig, ax = plt.subplots(1, 2, figsize=(32, 32))
fig.subplots_adjust(hspace=0, wspace=0)

ax[0].xaxis.set_major_locator(plt.NullLocator())
ax[0].yaxis.set_major_locator(plt.NullLocator())
ax[0].set_title("Sem blur", fontsize=48)
ax[0].imshow(img)

ax[1].xaxis.set_major_locator(plt.NullLocator())
ax[1].yaxis.set_major_locator(plt.NullLocator())
ax[1].set_title("Com blur", fontsize=48)
ax[1].imshow(img_blur)
```

```python
circulos = cv2.HoughCircles(img_cinza, cv2.HOUGH_GRADIENT, dp=1.2, minDist=100)
if circulos is not None:
    circulos = np.round(circulos[0, :]).astype("int")
    for (x, y, r) in circulos:
        cv2.circle(img, (x, y), r, (0, 255, 0), 4)
        
circulos_blur = cv2.HoughCircles(img_cinza_blur, cv2.HOUGH_GRADIENT, dp=1.2, minDist=100)
if circulos_blur is not None:
    circulos_blur = np.round(circulos_blur[0, :]).astype("int")
    for (x, y, r) in circulos_blur:
        cv2.circle(img_blur, (x, y), r, (0, 255, 0), 4)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(32, 32))
fig.subplots_adjust(hspace=0, wspace=0)

ax[0].xaxis.set_major_locator(plt.NullLocator())
ax[0].yaxis.set_major_locator(plt.NullLocator())
ax[0].set_title("Sem blur - %d" % len(circulos), fontsize=48)
ax[0].imshow(img)

ax[1].xaxis.set_major_locator(plt.NullLocator())
ax[1].yaxis.set_major_locator(plt.NullLocator())
ax[1].set_title("Com blur - %d" % len(circulos_blur), fontsize=48)
ax[1].imshow(img_blur)
```

4 círculos é muito pouco. Vamos tunar os parâmetros e tentar encontrar mais círculos na imagem original:

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display

def detectaCirculos(dp=1.2, minDist=100, blur=False):
    img = cv2.imread('plot.png')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    if blur:
        img = cv2.medianBlur(img,7)
    
    circulos = cv2.HoughCircles(img_cinza, cv2.HOUGH_GRADIENT, dp=dp, minDist=minDist)
    if circulos is not None:
        circulos = np.round(circulos[0, :]).astype("int")
        for (x, y, r) in circulos:
            cv2.circle(img, (x, y), r, (0, 255, 0), 4)
    
    fig, ax = plt.subplots(1, 1, figsize=(32, 32))
    fig.subplots_adjust(hspace=0, wspace=0)

    ax.set_title("Circulos: %d" % (len(circulos)), fontsize=48)
    ax.imshow(img)
```

```python
c = interactive(detectaCirculos, dp=(1,3,0.1), minDist=(50, 200), blur=False)
display(c)
```
