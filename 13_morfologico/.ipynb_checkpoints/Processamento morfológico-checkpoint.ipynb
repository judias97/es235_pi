## Aula 13 - Processamento morfológico

Neste notebook iremos usar os métodos do OpenCV para processamento morfológico.

Vamos importar as bibliotecas necessárias:

```python
%matplotlib inline
import cv2 # OpenCV
import matplotlib.pyplot as plt # Matplotlib
import numpy as np # Numpy
```

Vamos carregar a nossa imagem de testes e definir os kernels que serão usados no notebook:

```python
img = cv2.imread('j.png', 0)
plt.imshow(img, cmap="gray")
```

```python
kernels = {
    'retangulo': cv2.getStructuringElement(cv2.MORPH_RECT,(5,5)),
    'elipse': cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5)),
    'cruz': cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
}
```

### Erosão

Para fazer a erosão usaremos o método [cv2.erode()](https://docs.opencv.org/trunk/d4/d86/group__imgproc__filter.html#gaeb1e0c1033e3f6b891a25d0511362aeb). Como argumento passaremos a imagem, o kernel que vamos usar, e a quantidade de vezes (iterações) que queremos usar o erode na imagem.

```python
imagem_erosao = cv2.erode(img,kernels['retangulo'],iterations = 1)
plt.imshow(imagem_erosao, cmap="gray")
```

Podemos, ainda, comparar os kernels e mudar o número de iterações:

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display

def iteracoes(i=1):
    
    fig, ax = plt.subplots(1, 3, figsize=(32, 32))
    fig.subplots_adjust(hspace=0, wspace=0)

    ax[0].set_title("Retângulo - %d iterações" % (i), fontsize=48)
    ax[0].imshow(cv2.erode(img,kernels['retangulo'],iterations = i), cmap='gray')
    
    ax[1].set_title("Elipse - %d iterações" % (i), fontsize=48)
    ax[1].imshow(cv2.erode(img,kernels['elipse'],iterations = i), cmap='gray')
    
    ax[2].set_title("Cruz - %d iterações" % (i), fontsize=48)
    ax[2].imshow(cv2.erode(img,kernels['cruz'],iterations = i), cmap='gray')
```

```python
c = interactive(iteracoes, i=(0,5))
display(c)
```

Para o dilate, usaremos o método [cv2.dilate()](https://docs.opencv.org/trunk/d4/d86/group__imgproc__filter.html#ga4ff0f3318642c4f469d0e11f242f3b6c). Os argumentos são iguais ao da erosão.

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display

def iteracoes(i=1):
    
    fig, ax = plt.subplots(1, 3, figsize=(32, 32))
    fig.subplots_adjust(hspace=0, wspace=0)

    ax[0].set_title("Retângulo - %d iterações" % (i), fontsize=48)
    ax[0].imshow(cv2.dilate(img,kernels['retangulo'],iterations = i), cmap='gray')
    
    ax[1].set_title("Elipse - %d iterações" % (i), fontsize=48)
    ax[1].imshow(cv2.dilate(img,kernels['elipse'],iterations = i), cmap='gray')
    
    ax[2].set_title("Cruz - %d iterações" % (i), fontsize=48)
    ax[2].imshow(cv2.dilate(img,kernels['cruz'],iterations = i), cmap='gray')
```

### Opening

Para a abertura, usaremos o método [cv2.morphologyEx()](https://docs.opencv.org/trunk/d4/d86/group__imgproc__filter.html#ga67493776e3ad1a3df63883829375201f). Este método recebe além da imagem original, a morfologia a ser aplicada como uma flag e o kernel.

As flags disponíveis estão disponíveis na documentação. Para abertura, usamos a flag cv2.MORPH_OPEN.

```python
img = cv2.imread('j_opening.png', 0)
plt.imshow(img, cmap="gray")
```

```python
fig, ax = plt.subplots(1, 3, figsize=(32, 32))
fig.subplots_adjust(hspace=0, wspace=0)

ax[0].set_title("Retângulo", fontsize=48)
ax[0].imshow(cv2.morphologyEx(img, cv2.MORPH_OPEN, kernels['retangulo']), cmap='gray')
    
ax[1].set_title("Elipse", fontsize=48)
ax[1].imshow(cv2.morphologyEx(img, cv2.MORPH_OPEN, kernels['elipse']), cmap='gray')
    
ax[2].set_title("Cruz", fontsize=48)
ax[2].imshow(cv2.morphologyEx(img, cv2.MORPH_OPEN, kernels['cruz']), cmap='gray')
```

O top hat é altamente afetado pelo tamanho do kernel. Vamos alterar os valores e ver o comportamento:

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display

def kernel_size(i=1):
    
    kernels_tophat = {
    'retangulo': cv2.getStructuringElement(cv2.MORPH_RECT,(i,i)),
    'elipse': cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(i,i)),
    'cruz': cv2.getStructuringElement(cv2.MORPH_CROSS,(i,i))
}
    
    fig, ax = plt.subplots(1, 3, figsize=(32, 32))
    fig.subplots_adjust(hspace=0, wspace=0)

    ax[0].set_title("Retângulo - kernel %dx%d" % (i, i), fontsize=48)
    ax[0].imshow(cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernels_tophat['retangulo']), cmap='gray')
    
    ax[1].set_title("Elipse - kernel %dx%d" % (i, i), fontsize=48)
    ax[1].imshow(cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernels_tophat['elipse']), cmap='gray')
    
    ax[2].set_title("Cruz - kernel %dx%d" % (i, i), fontsize=48)
    ax[2].imshow(cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernels_tophat['cruz']), cmap='gray')
```

### Esqueletização

O processo de esqueletização é um pouco mais complicado. Vamos usar o algoritmo disponível [aqui](https://stackoverflow.com/questions/33095476/is-there-any-build-in-function-can-do-skeletonization-in-opencv), com alguns comentários e adaptações:

```python
img = cv2.imread('j.png', 0)
plt.imshow(img, cmap="gray")
```

```python
import cv2
import numpy as np

def skel(img):
    size = np.size(img)
    skel = np.zeros(img.shape,np.uint8)
    ret,img = cv2.threshold(img,127,255,0)
    element = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
    done = False

    while(not done):
        eroded = cv2.erode(img,element)
        temp = cv2.dilate(eroded,element)
        temp = cv2.subtract(img,temp)
        skel = cv2.bitwise_or(skel,temp)
        img = eroded.copy()

        zeros = size - cv2.countNonZero(img)
        if (zeros == size):
            done = True
    
    return skel
            
plt.imshow(skel(img), cmap="gray")
```

E se mudarmos o elemento estruturante?

```python
def skel(img, kernel):
    size = np.size(img)
    skel = np.zeros(img.shape,np.uint8)
    ret,img = cv2.threshold(img,127,255,0)
    element = kernel
    done = False

    while(not done):
        eroded = cv2.erode(img,element)
        temp = cv2.dilate(eroded,element)
        temp = cv2.subtract(img,temp)
        skel = cv2.bitwise_or(skel,temp)
        img = eroded.copy()

        zeros = size - cv2.countNonZero(img)
        if (zeros == size):
            done = True
    
    return skel
```

```python
fig, ax = plt.subplots(1, 3, figsize=(32, 32))
fig.subplots_adjust(hspace=0, wspace=0)

ax[0].set_title("Retângulo", fontsize=48)
ax[0].imshow(skel(img, kernels['retangulo']), cmap='gray')
    
ax[1].set_title("Elipse", fontsize=48)
ax[1].imshow(skel(img, kernels['elipse']), cmap='gray')
    
ax[2].set_title("Cruz", fontsize=48)
ax[2].imshow(skel(img, kernels['cruz']), cmap='gray')
```

### Exemplo: ressaltar contornos

Podemos subtrair a erosão da imagem original pra ter os contornos ressaltados. Faremos isso usando o mesmo código que usamos na parte de erosão, comparando os elementos estruturantes e as iterações:

```python
img = cv2.imread('j.png', 0)

from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display

def iteracoes(i=1):
    
    fig, ax = plt.subplots(1, 3, figsize=(32, 32))
    fig.subplots_adjust(hspace=0, wspace=0)

    ax[0].set_title("Retângulo - %d iterações" % (i), fontsize=48)
    ax[0].imshow(img - cv2.erode(img,kernels['retangulo'],iterations = i), cmap='gray')
    
    ax[1].set_title("Elipse - %d iterações" % (i), fontsize=48)
    ax[1].imshow(img - cv2.erode(img,kernels['elipse'],iterations = i), cmap='gray')
    
    ax[2].set_title("Cruz - %d iterações" % (i), fontsize=48)
    ax[2].imshow(img - cv2.erode(img,kernels['cruz'],iterations = i), cmap='gray')
```

```python
c = interactive(iteracoes, i=(0,5))
display(c)
```
